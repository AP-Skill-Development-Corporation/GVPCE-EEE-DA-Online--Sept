```python
# print the factors of a given number
n=int(input("enter any number"))
count=0
for i in range(1,n+1,1):
    if n%i==0:
        count=count+1
if count==2:
    print(n,"is prime")
else:
    print(n,"is not prime")
```

```python
# using while loop
n=int(input()) # 5times:1
#10:1,2,5
it=1
factors=0
while it<=n//2:
    if n%it==0:
        factors+=1
    it+=1
if factors==1:
    print('prime')
else:
    print('not prime')
```

```python
# prime numbers in a range
lw,up=int(input()),int(input())
mx=0
for val in range(lw,up+1):
    count=0
    for dig in range(1,val+1):
        if val%dig==0:
            count+=1
    if count==2:
        print(val,end=" ")
        mx+=1
print("\nno.of primes",mx)
```

```python
a=int(input("Enter starting number:"))
b=int(input("enter ending number"))
print("the prime numbers between",a,"and",b,"is")
for i in range(a+1,b+1,1):
    count=0
    for k in range(1,i+1,1):
        if i%k==0:
            count+=1
    if count==2:
        print(i)
    
```

```python
#using while loop
lw,up=int(input()),int(input())
while lw<=up:
    count,it=0,1
    while it<=lw:
        if lw%it==0:
            count+=1
        it+=1
    if count==2:
        print(lw,end=" ")
    lw+=1
```

        - defined by the user

```python
# def is_prime
def is_prime(n):
    count=0
    for dig in range(1,n//2+1):
        if n%dig==0:
            count+=1
    if count==1:
        return True
    else:
        return False
    
is_prime(14)
```

```python
# def is_prime
def is_prime(n):
    count=0
    for dig in range(1,n//2+1):
        if n%dig==0:
            count+=1
    if count==1:
        return True
    else:
        return False
    
def prime_range(lw,up):
    for val in range(lw,up+1):
        if is_prime(val):
            print(val,end=" ")
            
prime_range(100,200)
```
