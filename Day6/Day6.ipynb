##### sample plot

```python
x=[1,2,3,4,5]
y=[10,20,30,40,50] # x&y are continuous
z=[34.5,67.8,56.7,34.5,23.6]
plt.plot(x,y)
plt.show()
```

```python
# lables
plt.plot(x,z)
plt.xlabel('x-axis')
plt.ylabel('z points')
plt.show()
```

```python
#adding labels in the plot itself
plt.plot(x,y,label='first')
plt.plot(x,z,label='second')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("Multiple graphs")
plt.show()
```

```python
# giving axis range
li=[1,2,3,4,5,6] # 10 diff
li2=[10.5,11.5,12.5,13.5,14.5,15.5] # 
plt.plot(li,li2)
plt.axis([0,10,0,20]) # x-->[0,10],y-->[0,20.0]
plt.title("sample3 graph")
plt.show()
```

```python
# empty graph
plt.title("hi")
```

```python
# single point(x,y)
list=[1,2,3,4,5,6]
list2=[10,15,26,37,49,55]
plt.plot(list,list2)
plt.axis([0,10,10,100])
plt.title("sample3 graph")
plt.show()
```

```python
# attributes
plt.plot(x,y,color='red',linestyle='dashed')
plt.show()
```

```python
plt.plot(li,li2,color='b',linestyle='--',marker='*')
plt.show()
```

```python
plt.plot(x,z,color='g',linestyle='--',marker='o',linewidth=3,
        markersize=8)
plt.show()
```

```python
from matplotlib import style
style.use('ggplot') # grid plot
days=[1,2,3,4,5]
temp=[36,35,37,40,38.3]
plt.xlabel('days',fontsize=15)
plt.ylabel('temperature',fontsize=10)
plt.plot(days,temp,color='black',marker='o',linewidth=3)
plt.show()
```

```python
vizag=[34.5,33.7,35.2,37.5,40.4]
plt.plot(days,vizag,"m*--",linewidth=3,
         label='vizag temp') #color,marker,linestyle
plt.plot(days,temp,color='g',marker='o',linewidth=4,label='gnt')
plt.title('Gnt Temp vs Vizag Temp ')
plt.legend() # 1,2,3,4
plt.show()
```

```python
vizag=[34.5,33.7,35.2,37.5,40.4]
plt.plot(days,vizag,"m*--",linewidth=3,
         label='vizag temp') #color,marker,linestyle
plt.plot(days,temp,color='g',marker='o',linewidth=4,label='gnt')
plt.title('Gnt Temp vs Vizag Temp ')
plt.legend(loc=4) # 1,2,3,4
plt.show()
```

```python
vizag=[34.5,33.7,35.2,37.5,40.4]
plt.plot(days,vizag,"m*--",linewidth=3,
         label='vizag temp') #color,marker,linestyle
plt.plot(days,temp,color='g',marker='o',linewidth=4,label='gnt')
plt.title('Gnt Temp vs Vizag Temp ')
plt.legend() # 1,2,3,4
plt.grid(color='r',linestyle='-',linewidth=2) # adding grid attributes
plt.show()
```

#### histogram
- represented as bins
- plt.hist(ele,bins=count)

```python
rn=np.random.randint(15,45,100) # 100 values==40
rn2=np.random.randint(10,40,100)
plt.hist(rn,bins=40)
plt.show()
```

```python
plt.hist(rn2,bins=50)
plt.show()
```

```python
plt.hist(rn2,bins=100)
plt.show()
```

```python
plt.hist(rn,color='c',orientation='horizontal')
plt.show()
```

```python
# multiple histograms in a plane
plt.figure(figsize=(8,5))
vals=[num for num in range(10,50,5)] # 10,15,20,25..50
plt.hist([rn,rn2],vals,histtype='bar',color=['m','y'])
plt.show()
```

#### pie plot
- cicular plot

```python
nums=[56,25,30,24] # 100% 4 partitions
plt.pie(nums)
plt.show()
```

```python
vals=[50,80,34,56,12,90,56] # len(it) regions
plt.pie(vals)
plt.show()
```

```python

fbk=['average','above average','good','excellent']
plt.pie(nums,labels=fbk)
plt.show()
```

```python
plt.pie(nums,labels=fbk,colors=['cyan','red','green','pink'])
plt.show()
```

```python
# adding % of region
plt.pie(nums,labels=fbk,colors=['cyan','red','green','pink'],
       autopct="%f.%%")
plt.title("sorting by responses")
plt.show()
```

```python
# adding % of region
ex=[0,0,0,1]
plt.pie(nums,labels=fbk,colors=['cyan','red','green','pink'],
       autopct="%f.%%",shadow=True,startangle=180,
       explode=ex)
plt.title("sorting by responses")
plt.show()
```

#### stackplot
- one by one arranging
- Last In First Out

```python
x=[9,3,5,4,8]
y=[9,4,5,6,2]
z=[6,8,2,3,1]
n=np.array([8,4,9,4,2])
plt.stackplot(x,y,z,n)
plt.show()
```

```python
help(plt.stackplot)
```

```python
f=[1,2,3,4,5]
s={'y1':[5,6,4,5,7],'y2':[1,6,4,5,6],'y3':[1,1,2,3,1]}
plt.stackplot(f,s.values())
plt.legend(loc=3)
plt.title('stack plot')
plt.show()
```
